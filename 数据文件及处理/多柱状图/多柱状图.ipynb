```python
import csv

result = []

# 遍历2014年到2022年的大规模枪击案数据文件
for year in range(2014, 2023):
    filename = f"{year}大规模.csv"

    with open(filename, 'r', newline='') as file:
        reader = csv.reader(file)
        next(reader)  
        state_count = {}
        for row in reader:
            state = row[2] 
            state_count[state] = state_count.get(state, 0) + 1
        for state, count in sorted(state_count.items()):
            result.append("{{ Year: {}, State: '{}', Count: {} }}".format(year, state, count))

for item in result:
    print(item,end=',\n')
```
